```python


# Define symptom keyword lists
speech_symptoms = [
    "speech", "talk", "language", "communication", "slur", "slurred", 
    "confusion", "understand", "difficulty speaking", "trouble speaking"
]

paralysis_symptoms = [
    "weakness", "numbness", "paralysis", "arm", "leg", "face drooping", 
    "hand", "one side", "half body", "unable to move", "immobility"
]

vision_symptoms = [
    "vision", "eyes", "blur", "blind", "eye sight", "see", "double vision"
]

balance_symptoms = [
    "dizzy", "balance", "coordination", "walking", "staggering", "vertigo"
]

headache_symptoms = [
    "headache", "pain", "severe head", "migraine"
]

unaware_responses = [
    "don't know", "no idea", "na", "nan", "none", "not sure"
]

# Cleaning function
def clean_text(text):
    if pd.isna(text):
        return "unknown"
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Multi-label categorization function
def categorize_symptom_multi(resp):
    resp = clean_text(resp)
    
    if resp in unaware_responses or resp == "unknown":
        return "Unaware"
    
    categories = []
    if any(word in resp for word in speech_symptoms):
        categories.append("Speech / Confusion")
    if any(word in resp for word in paralysis_symptoms):
        categories.append("Paralysis / Weakness")
    if any(word in resp for word in vision_symptoms):
        categories.append("Vision Problems")
    if any(word in resp for word in balance_symptoms):
        categories.append("Balance / Dizziness")
    if any(word in resp for word in headache_symptoms):
        categories.append("Headache")
    
    if not categories:
        if len(resp.split()) > 6:
            categories.append("Multiple Symptoms / Aware")
        else:
            categories.append("General Awareness")
    
    return ", ".join(categories)

# Apply categorization
col = "what_are_the_symptoms_of_brain_stroke?"
df["stroke_symptoms"] = df[col].apply(categorize_symptom_multi)

# View distribution
print(df["stroke_symptoms"].value_counts())
```

```python
# High blood pressure / heart-related
bp_heart = [
    "high blood pressure", "hypertension", "heart disease", "heart diseases",
    "atrial fibrillation", "high red blood cell count"
]

# Diabetes / Obesity / Metabolic
diabetes_obesity = [
    "diabetes", "obesity", "overweight", "high cholesterol", "high cholesterol and lipids",
    "poor diet", "physical inactivity", "sedentary", "stressful living"
]

# Smoking / Alcohol / Drugs
smoking_alcohol = [
    "smoking", "excessive alcohol", "alcohol consumption", "addictive drug", "cocaine", "binge drinking"
]

# Age / Family / Gender
age_family = [
    "family history of stroke", "prior stroke", "age", "older adults", "men", "gender"
]

# Others / Unknown
other_unknown = [
    "nan", "no idea", "don't know", "na", "none", "not sure", "unknown"
]


def clean_text(text):
    if pd.isna(text):
        return "unknown"
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["cleaned_risk"] = df["what_are_the_risk_factors_for_brain_stroke?"].apply(clean_text)
def categorize_risk(resp):
    resp = str(resp).lower()
    categories = []
    
    if any(word in resp for word in bp_heart):
        categories.append("BP / Heart")
    if any(word in resp for word in diabetes_obesity):
        categories.append("Diabetes / Obesity")
    if any(word in resp for word in smoking_alcohol):
        categories.append("Smoking / Alcohol / Drugs")
    if any(word in resp for word in age_family):
        categories.append("Age / Family / Gender")
    if any(word in resp for word in other_unknown):  
        categories.append("Unaware")
    if resp == "unknown":
        categories.append("no response")
    if not categories:
        categories.append("Other")
    
    return ", ".join(categories)

df["risk_factors"] = df["cleaned_risk"].apply(categorize_risk)
print(df["risk_factors"].value_counts())
```

```python

advice_mapping = {
    "emergency": ["call emergency", "911", "999", "112", "emergency services"],
    "medical_checkup": ["go for checkup", "medical checkup", "consult doctor", "take medicines", "consultation", "medical treatment", "checkup"],
    "calm_reassure": ["stay calm", "keep calm", "reassure", "dont panic", "do not panic", "calm down", "keep person calm", "keep them calm"],
    "no_food_drink_sleep": ["do not give food", "do not give drink", "dont sleep", "do not sleep", "dont eat or drink", "do not offer food or drink", "no food or drink"],
    "observation_monitor": ["note time", "note symptoms", "monitor", "keep an eye on symptoms", "observe", "stay with person", "keep them comfortable"],
    "rest_safety": ["take rest", "keep lying down", "safe position", "keep them safe", "do not drive"],
    "unknown": ["no idea", "dont know", "na", "not aware", "unknown", "cannot suggest", "naa"]
}

def map_advice(text):
    # Handle NaN or empty strings first
    if pd.isna(text) or str(text).strip().lower() in ["nan", ""]:
        return "No response"
    
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', ' ', text)  # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # normalize spaces
    
    for category, keywords in advice_mapping.items():
        for kw in keywords:
            if kw in text:
                return category
    return "other"

# Apply mapping
df["what_advice_would_you_give_for_someone_experiencing_stroke_symptoms"] = df["what_advice_would_you_give_if_someone_you_know_experiences_symptoms_of_a_brain_stroke?"].apply(map_advice)

# Check results
print(df["what_advice_would_you_give_for_someone_experiencing_stroke_symptoms"].value_counts())
```

```python


# Sample mapping function
def clean_salary(sal):
    if pd.isna(sal) or str(sal).strip().lower() in ['nan', '']:
        return "Not Specified"
    
    sal = str(sal).replace("₹", "").replace(",", "").strip()  # remove ₹ and commas
    sal = re.sub(r'\s+', '', sal)  # remove extra spaces
    
    # Handle "more than" and "less than"
    if "morethan" in sal.lower():
        return "Above 8,00,000"
    elif "lessthan" in sal.lower():
        return "Below 2,00,000"
    
    # Match ranges using regex
    match = re.match(r'(\d+)-(\d+)', sal)
    if match:
        low = int(match.group(1))
        high = int(match.group(2))
        
        if low < 200001:
            return "Below 2,00,000"
        elif 200001 <= low <= 400000:
            return "2,00,001 - 4,00,000"
        elif 400001 <= low <= 600000:
            return "4,00,001 - 6,00,000"
        elif 600001 <= low <= 800000:
            return "6,00,001 - 8,00,000"
        else:
            return "Above 8,00,000"
    
    # Default for any unmatched cases
    
df["salary"] = df["salary"].apply(clean_salary)

# Check results
print(df["salary"].value_counts())
```

```python

def standardize_age(value):
    if pd.isna(value):
        return "Unknown"
    value = str(value).strip()
    
    # Handle ranges
    if '-' in value:
        start, end = value.split('-')
        start, end = int(start), int(end)
        return f"{start}-{end}"
    
    # Handle 'above of 60'
    if 'above' in value.lower():
        return '60+'
    
    # Handle numeric values
    try:
        num = float(value)
        if num <= 15:
            return '0-15'
        elif num <= 25:
            return '15-25'
        elif num <= 40:
            return '26-40'
        elif num <= 60:
            return '41-60'
        else:
            return '60+'
    except:
        return "Other"
df['age'] = df['age'].apply(standardize_age)
df['age'].value_counts()

```
